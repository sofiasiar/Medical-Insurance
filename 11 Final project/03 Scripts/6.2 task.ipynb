## Importing Libraries

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os
```

```python
# Importing dataframes
path= r"C:\Users\sosia\11 Final project"
df = pd.read_csv(r'C:\Users\sosia\OneDrive\Desktop\CF\Data immersion\6\6.1\archive\insurance.csv', index_col = False)
```

## Create a correlation heatmap

```python
df.columns
```

```python
# Create a subset excluding the "Date" and "No" columns

sub = df[['age', 'bmi', 'children','charges']]
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

# Create a Scatterplot

```python
# Create a scatterplot for the "house age" and "price per unit" columns in seaborn

sns.lmplot(x = 'age', y = 'charges', data = df)
```

```python
# export 
sns.savefig("scatterplot.png", dpi=300, bbox_inches='tight')
```

#### The scatterplot shows a positive correlation between age and charges.

## Create a pair plot

```python
# Create a pair plot 

g = sns.pairplot(sub)
```

```python
g.savefig("pairplot.jpg", dpi=300)
```

#### BMI and Charges: While the heatmap does not show a very strong correlation here we observe a strong positive correlation 
#### This suggests that as BMI increases, charges tend to increase as well. 
#### The scatter plot of BMI vs. Charges visually confirms this, showing an upward trend.  
#### Individuals with higher BMIs are likely to incur higher medical charges. 

#### Age and Charges: A moderate positive correlation is observed. 

#### Potential Outliers in BMI:  While BMI seems roughly normally distributed, there are a few 
#### individuals with very high BMIs.

## Create a categorical plot

```python
# Use a histogram to visualize the distribution of the variables. 

sns.histplot(df['charges'], bins = 20, kde = True)
```

```python
df.loc[df['charges'] < 10000, 'Price category'] = 'Low price'
```

```python
df.loc[(df['charges'] >= 10000) & (df['charges'] <25000), 'Price category'] = 'Middle price'
```

```python
df.loc[df['charges'] >= 25000, 'Price category'] = 'High price'
```

```python
df['Price category'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="smoker", y="charges", hue="Price category", data=df)
```

```python
g.savefig("categorical_plot1.png", dpi=300, bbox_inches="tight")
```

#### Charges and smoking are related since we can observe that high prices are more prevalent among smokers

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="sex", y="charges", hue="Price category", data=df)
```
